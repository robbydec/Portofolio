```python
df.sort_values(by=['categori'], inplace=False, ascending=False)
```

```python
from matplotlib import rcParams
rcParams['figure.figsize'] = 20,10
df['tanggal'] = pd.to_datetime(df['tanggal'])
df.plot(x='tanggal', y= ['pm10','so2','co','o3','no2'])
plt.title('Tingkat Kandungan Udara DKI Januari 2021')
```

```python
dflist = ['pm10','so2','co','o3','no2']
df['total per hari'] = df[dflist].sum(axis=1)
df.drop(['max','categori','location'], axis=1, inplace=True)
df
```

```python
df.plot(x='tanggal', y= ['total per hari'])
plt.title('Tingkat Polusi Rata-rata DKI Januari 2021')
```

```python
dflist1 = ['pm10','so2','co','o3','no2','total per hari']
df.loc['total'] = df[dflist1].sum(axis=0)
df.tail()
```

```python
df.loc['total', :]
df.loc['%'] = (df.loc['total']/df.loc['total', 'total per hari'].sum())*100
df.tail()
```

```python
total_persen = df.loc['%']
total_persen
```

```python
total_persen_chart = total_persen['pm10' : "no2"]
total_persen_chart
```

```python
total_persen_chart.plot(kind = 'pie', autopct='%1.1f%%')
plt.title('Tingkat Kandungan Persen')
```
